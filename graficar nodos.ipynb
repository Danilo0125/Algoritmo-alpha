```python
import networkx as nx
import matplotlib.pyplot as plt

# Crear un nuevo grafo dirigido
G = nx.DiGraph()

# Agregar nodos
G.add_node('il')
G.add_node('A')
G.add_node('Ol')

# Agregar bordes/aristas
G.add_edge('il', 'A')
G.add_edge('A', 'Ol')

# Posiciones de los nodos
pos = nx.spring_layout(G)

# Dibujar nodos con diferentes formas
nx.draw_networkx_nodes(G, pos, nodelist=['il'], node_shape='o', node_color='skyblue', node_size=1000)
nx.draw_networkx_nodes(G, pos, nodelist=['A'], node_shape='s', node_color='lightgreen', node_size=1000)
nx.draw_networkx_nodes(G, pos, nodelist=['Ol'], node_shape='o', node_color='lightcoral', node_size=1000)

# Dibujar bordes/aristas con flechas
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=40)

# Dibujar etiquetas de los nodos
nx.draw_networkx_labels(G, pos, font_size=12, font_color='black', font_weight='bold')

# Mostrar el gráfico
plt.show()
```

```python
# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos y aristas con etiquetas
edges = [('IL','A',"IL->A"), ('IL','B',"IL->B"), ('D','OL',"D->OL"),('B', "P({B},{D})", "B→P({B},{D})"), ('A', "P({B},{D})", "A→P({B},{D})"), ("P({B},{D})", 'D', "P({B},{D})→D")]
G.add_edges_from([(u, v) for u, v, _ in edges])

# Posición de los nodos
pos = nx.spring_layout(G)

# Dibujar nodos redondos
nx.draw_networkx_nodes(G, pos, nodelist=['IL', 'OL', 'P({B},{D})'], node_shape='o', node_color='lightblue', node_size=500)

# Dibujar nodos cuadrados
nx.draw_networkx_nodes(G, pos, nodelist=['A', 'B', 'D'], node_shape='s', node_color='lightgreen', node_size=300)

# Dibujar aristas con flechas
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=25, edge_color="black")

# Dibujar etiquetas de nodos con tamaño de fuente más pequeño
nx.draw_networkx_labels(G, pos, font_size=8, font_weight="bold")

# Diccionario con etiquetas de aristas
edge_labels = {(u, v): label for u, v, label in edges}

# Dibujar etiquetas de las aristas con tamaño de fuente más pequeño
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, font_color="red")

# Mostrar el gráfico
plt.show()
```

```python
# Crear un grafo dirigido
G = nx.DiGraph()

# Agregar nodos y aristas
edges = [('IL','A'), ('IL','B'), ('D','OL'),('B', "P({B},{D})"), ('A', "P({B},{D})"), ("P({B},{D})", 'D')]
G.add_edges_from([(u, v) for u, v in edges])

# Posición de los nodos
pos = nx.spring_layout(G)

# Dibujar nodos redondos
nx.draw_networkx_nodes(G, pos, nodelist=['IL', 'OL', 'P({B},{D})'], node_shape='o', node_color='lightblue', node_size=500)

# Dibujar nodos cuadrados
nx.draw_networkx_nodes(G, pos, nodelist=['A', 'B', 'D'], node_shape='s', node_color='lightgreen', node_size=300)

# Dibujar aristas con flechas
nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=25, edge_color="black")

# Dibujar etiquetas de nodos con tamaño de fuente más pequeño
nx.draw_networkx_labels(G, pos, font_size=8, font_weight="bold")

# Mostrar el gráfico
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

# Parámetros para el corazón
t = np.linspace(0, 2 * np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)

# Graficar el corazón
plt.figure(figsize=(8, 8))
plt.plot(x, y, color='red')
plt.title('Gráfico de un corazón')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axis('equal')
plt.show()
```
